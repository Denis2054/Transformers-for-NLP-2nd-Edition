```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate a random dataset with 1000 data points
data = [random.randint(1, 1000) for i in range(1000)]

# Convert the data to a numpy array
data = np.array(data)

# Reshape the data to have a single feature
data = data.reshape(-1, 1)

# Define the number of clusters
n_clusters = 6

# Apply k-means clustering to the data
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(data)

# Get the labels for each data point
labels = kmeans.predict(data)

# Get the centroids of each cluster
centroids = kmeans.cluster_centers_

# Plot the data points and centroids
colors = ['b', 'g', 'r', 'c', 'm', 'y']
for i in range(n_clusters):
    plt.scatter(data[labels == i], np.zeros_like(data[labels == i]), color=colors[i])
plt.scatter(centroids, np.zeros(n_clusters), marker='x', s=200, linewidths=3, color='k')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = pd.read_csv('iris.data', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

# Extract the features from the dataset
X = iris.iloc[:, :-1].values

# Define the number of clusters
n_clusters = 3

# Apply k-means clustering to the data
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(X)

# Get the labels for each data point
labels = kmeans.predict(X)

# Get the centroids of each cluster
centroids = kmeans.cluster_centers_

# Plot the data points and centroids
colors = ['b', 'g', 'r']
for i in range(n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], color=colors[i])
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='k')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('K-Means Clustering on Iris Dataset')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import requests

# Download the Iris dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
response = requests.get(url)
open('iris.csv', 'w').write(response.text)

# Load the Iris dataset
iris = pd.read_csv('iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

# Extract the features from the dataset
X = iris.iloc[:, :-1].values

# Define the number of clusters
n_clusters = 3

# Apply k-means clustering to the data
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(X)

# Get the labels for each data point
labels = kmeans.predict(X)

# Get the centroids of each cluster
centroids = kmeans.cluster_centers_

# Plot the data points and centroids
colors = ['b', 'g', 'r']
for i in range(n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], color=colors[i])
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='k')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('K-Means Clustering on Iris Dataset')
plt.show()
```
